```python
!pip install diffusers transformers accelerate safetensors gradio --quiet

import torch
from diffusers import AutoPipelineForText2Image
import gradio as gr

device = "cpu"
print("Using:", device)

pipe = AutoPipelineForText2Image.from_pretrained(
    "stabilityai/sd-turbo",
    torch_dtype=torch.float32
).to(device)

def generate(prompt):
    image = pipe(prompt, num_inference_steps=2, guidance_scale=0.0).images[0]
    return image

iface = gr.Interface(
    fn=generate,
    inputs="text",
    outputs="image",
    title="Stable Diffusion Turbo (CPU Optimized)",
    description="Ultra-fast distilled Stable Diffusion model for CPU."
)

iface.launch()
```
